## Importando o arquivo CSV para o notebook.


```python
import pandas as pd

df = pd.read_csv('/content/Salary_dataset.csv')
display(df.head())
```

## Analisando os dados para entender sua estrutura e conteúdo.


```python
df.info()
display(df.describe())
display(df.isnull().sum())
```

## Preparando os dados

Separarando as variáveis independentes (anos de experiência) da variável dependente (salário).


```python
X = df[['YearsExperience']]
y = df['Salary']

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

## Dividindo os dados

Separando os dados em conjuntos de treino e teste.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Criando e treinando o modelo

Usando a Regressão Linear para criar e treinar o modelo com os dados de treino.


```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

## Avaliando o modelo

Verificando o desempenho do modelo com os dados de teste.


```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"O R² score é: {r2:.2f}")
print(f"O Erro Absoluto Médio (MAE) é: {mae:.2f}")
print(f"O Erro Quadrático Médio (MSE) é: {mse:.2f}")
```

## Fazer predições

Usar o modelo treinado para prever salários.


```python
y_pred = model.predict(X_test)
```

## Visualizando os resultados

Apresentando os resultados de forma clara e visual.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, color='blue', label='Dados de Treino')
plt.plot(X_train, model.predict(X_train), color='red', label='Linha de Regressão (Treino)')
plt.title('Salário vs. Anos de Experiência (Treino)')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X_test, y_test, color='green', label='Dados de Teste')
plt.plot(X_test, y_pred, color='red', label='Linha de Regressão (Teste)')
plt.title('Salário vs. Anos de Experiência (Teste)')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
plt.legend()

plt.tight_layout()
plt.show()
```
